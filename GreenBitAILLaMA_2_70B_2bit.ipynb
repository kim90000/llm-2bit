```python
import torch
from transformers import AutoTokenizer, GenerationConfig
from auto_gptq import AutoGPTQForCausalLM

# تحديد مسار مجلد النموذج
MODEL_DIR = "/content/low_bit_llama/cache/1"

# تحميل المحول اللغوي
tokenizer = AutoTokenizer.from_pretrained(MODEL_DIR)

# تحميل النموذج
model = AutoGPTQForCausalLM.from_quantized(
    MODEL_DIR,
    device="cuda" if torch.cuda.is_available() else "cpu",
    use_safetensors=True,
    trust_remote_code=True
)

# تحميل إعدادات التوليد من الملف إذا كان موجودًا
try:
    gen_config = GenerationConfig.from_pretrained(MODEL_DIR)
except:
    gen_config = GenerationConfig(max_length=50, temperature=0.7, top_p=0.9)

# إدخال نص للاختبار
prompt = "Who is Einstein?"
inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

# تشغيل الاستدلال باستخدام الإعدادات المحملة
with torch.no_grad():
    output = model.generate(**inputs, generation_config=gen_config)

# طباعة الإخراج
print(tokenizer.decode(output[0], skip_special_tokens=True))
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch
MODEL_DIR = "/content/low_bit_llama/cache/1"

model = AutoModelForCausalLM.from_pretrained(MODEL_DIR, device_map="auto")
tokenizer = AutoTokenizer.from_pretrained(MODEL_DIR)

# إدخال نص للاختبار
prompt = "hi"

# تجهيز الإدخال باستخدام المحول اللغوي
inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

# تشغيل الاستدلال
with torch.no_grad():
    output = model.generate(**inputs, max_length=5)  # تحديد طول النص المولد

# طباعة الإخراج
print(tokenizer.decode(output[0], skip_special_tokens=True))
```

```python
if "CHAT" in args.model_size:
    prompt = '''You are a helpful assistant, please tell me the meaning of life.'''
else:
    prompt = '''The meaning of life is'''

batch = tokenizer(prompt, return_tensors="pt", add_special_tokens=False)
batch = {k: v.cuda() for k, v in batch.items()}
model.cuda()

for i in range(10):
    with torch.no_grad():
        generated = model.generate(
            inputs=batch["input_ids"],
            do_sample=True,
            use_cache=True,
            repetition_penalty=1.5,
            max_new_tokens=10,
            temperature=0.9,
            top_p=0.95,
            top_k=20,
            return_dict_in_generate=True,
            output_attentions=False,
            output_hidden_states=False,
            output_scores=False
        )
    result_text = tokenizer.decode(generated['sequences'].cpu().tolist()[0])
    print(result_text + "\n")
```
