```python
from transformers import pipeline
import torch
question = "Who is Einstein?"
generator = pipeline("text-generation", model="GreenBitAI/yi-6b-chat-w4a16g128", torch_dtype=torch.float16, device_map="auto")
output = generator([{"role": "user", "content": question}], max_new_tokens=12, return_full_text=False)[0]
print(output["generated_text"])
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
tokenizer = AutoTokenizer.from_pretrained("GreenBitAI/yi-6b-chat-w4a16g128")
model = AutoModelForCausalLM.from_pretrained("GreenBitAI/yi-6b-chat-w4a16g128", trust_remote_code=True, torch_dtype=torch.float16, device_map="auto")

# we use alpaca prompt
input_content = "hi"
input_ids = tokenizer.encode(input_content, return_tensors="pt")
output = model.generate(input_ids, max_length=11, temperature=0.7)
output_text = tokenizer.decode(output[0], skip_special_tokens=True)
print(output_text)
```
